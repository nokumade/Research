# Making dummy variables from categorical values for segmentation

```python
df2.BANKACCOUNT_FLAG.replace([0, 1, 2, 3, 4], [0, 1, 2, 2, 2], inplace=True)
df2.SEX.replace(['мужской', 'женский'], [1, 0], inplace=True)
df3 = pd.get_dummies(df2, columns=cat_feat_sign, drop_first=True)
```

## Part II

# Elbow method

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
inertia = []
for x in range(1, 13):
  kmeans = KMeans(n_clusters=x, random_state=0, n_init="auto").fit(df3)
  inertia.append(kmeans.inertia_)
```

```python
!pip install kneed

from kneed import KneeLocator # used in order to find the best number of clusters mathematically
```

```python
kneedle = KneeLocator(range(1, 13), inertia, curve='convex', direction='decreasing')
optimal_clusters = kneedle.elbow

print(f"The optimal number of clusters is: {optimal_clusters}")
```

```python
plt.plot(inertia)
```

# But we will take 5 clusters, since it is the most popular choice when it comes to choosing the number of clusters.

# Visualising the clusters.

```python
kmeans = KMeans(n_clusters=5, random_state=0, n_init="auto")
df3['Clusters'] = kmeans.fit_predict(df3)
```

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df3.drop('Clusters', axis=1))
```

```python
plt.scatter(pca_result[:,0],pca_result[:,1], c=df3['Clusters'])
plt.show()
```

```python
cl1 = df3[df3['Clusters'] == 0]
cl2 = df3[df3['Clusters'] == 1]
cl3 = df3[df3['Clusters'] == 2]
cl4 = df3[df3['Clusters'] == 3]
cl5 = df3[df3['Clusters'] == 4]
```

## RFM

```python
rfm = df3.copy()
```

```python
rfm = rfm[['sum_of_paym_months', 'all_credits', 'min_MnthAfterLoan']]
rfm
```

```python
rfm = rfm.rename(columns={'sum_of_paym_months': 'Monetary',
                    'all_credits': 'Frequency',
                    'min_MnthAfterLoan': 'Recency'})
```

```python
r = pd.qcut(rfm['Recency'], q=5, labels=range(5, 0, -1))
f = pd.qcut(rfm['Frequency'], q=5, labels=range(1,6))
m = pd.qcut(rfm['Monetary'], q=5, labels=range(1,6))
rfm = rfm.assign(R=r.values, F=f.values, M=m.values)
rfm['rfm_group'] = rfm[['R','F','M']].apply(lambda v: ''.join(v.astype(str)), axis=1)
rfm['total_rfm_score'] = rfm[['R','F','M']].sum(axis=1)
rfm2 = rfm.sort_values('total_rfm_score')
```

# Plotting the total RFM against frequency

```python
#plt.figure(figsize=(15,15))
plt.xticks(rotation=90)
plt.xlabel("Total RFM Score")
plt.ylabel("Frequency")
plt.hist(rfm2['total_rfm_score'], bins=13)
plt.show()
```
