4. Forecasting models
AR(p) MODEL (AutoRegressive Model of Order p)
In an autoregressive model, the future behaviour of a variable is predicted by considering a linear combination of its past values. Essentially, this means that the current state of the variable y(t) is determined as follows:

y(t) = C + Σ(i=1 to p) φ(i) * y(t-i) + ε(t) or =f(y_t-1, y_t-2, …, ε(t)),

where "p" is a parameter that can vary in value, "C" represents a constant, φ(i) are coefficients that determine the influence of the previous p values of y(t), and ε(t) is the error term at time t.

MA(q) MODEL (Moving Average Model of Order q)
The moving average model is based on a function of error terms and uses early errors terms for prediction. It is expressed as:

y(t) = μ + Σ(i=1 to q) θ(i) * ε(t-i) + ε(t),

where y(t) represents the observed data point at time t, "μ" is the mean of the time series, “θ(i)” are coefficients that determine the influence of past q error terms ε(t-i), and ε(t) is the error term at time t. Fitting the MA model can be more challenging compared to the AR model due to the need to estimate the coefficients θ(i).
This model is harder than AR model to fit the time series because the errors are not foreseeable.

ARMA(p, q) MODEL (AutoRegressive Moving Average Model)
The ARMA model combines both autoregressive (AR) and moving average (MA) components. It is expressed as:

y(t) = c + Σ(i=1 to p) φ(i) * y(t-i) + Σ(j=1 to q) θ(j) * ε(t-j) + ε(t)

In this model, y(t) is the observed data point at time t, "c" is a constant, φ(i) are autoregressive coefficients, θ(j) are moving average coefficients, and ε(t) represents the error term at time t. The ARMA model allows for capturing both the autocorrelation of the time series (AR) and the influence of past error terms (MA).

These models are essential tools in time series analysis and forecasting, helping analysts understand and predict patterns in time-dependent data.

ARIMA MODEL (AutoRegressive Integrated Moving Average)

Compared to ARMA, the ARIMA model is suitable for handling non-stationary time series behaviour. This model transforms non-stationary data into a stationary format through differencing once or more times.

The ARIMA(p, d, q) model is an AutoRegressive Integrated Moving Average model, where "yt" is modelled as:

yt = c + φ1yt-1 + φ2yt-2 + ... + φpyt-p + θ1εt-1 + θ2εt-2 + … + θqεt + εt
ŷt=μ+ϕ1yt-1+…+ϕpyt-p-θ1et-1-…- θqet-q

Here, "yt" represents a different time series.

The ARIMA model involves three parameters: "p" for the order of the AutoRegressive (AR) part, "d" for the degree of differencing, and "q" for the order of the Moving Average (MA) part.

The model combines three fundamental methods:

AR (AutoRegressive) Part: This part regresses "yt" with its own lagged values.

Differencing Part: Differencing transforms the non-stationary time series into a stationary model. For example, first-order differencing is expressed as D(i) = Data(i) - Data(i-1). The parameter "d" represents the number of times the differencing method was applied.

MA (Moving Average) Part: The MA part considers the number of lagged variables used in the model to account for the influence of past error terms.

In summary, ARIMA is a versatile model that combines these three components to handle non-stationary time series data effectively. The "Integrated" component (denoted as "I" for differencing) makes ARIMA suitable for transforming non-stationary data into a stationary format before applying ARMA modelling, especially when dealing with data exhibiting trends or seasonality. ARMA, in contrast, assumes that the data is already stationary and does not include the differencing component.

# Dickey Fuller Test

Nifty

```python
"""
The Augmented Dickey-Fuller Test test whether the time-series is non-stationary, so the variance is not constant over time.
Our Null Hypothesis is that the Time Series is non-stationary and Alternative Hypothesis is the oposite. We test on a 5% significance level.
Before differencing the Open column we got that the p-value is >0.43, so we failed to reject the H0 because in 43% of time under the H0 we would get the same results.
Therefore we decided to difference our data and apply the hypothesis test.
"""


def adf_test(series,title=''):
    """
    Pass in a time series and an optional title, returns an ADF report
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC')

    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)

    for key,val in result[4].items():
        out[f'critical value ({key})']=val

    print(out.to_string(), '\n')

    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("Data has a unit root and is non-stationary \n\n")

    return out

adf_test(df_dated['Open'],title='Open Stock Data');
df_dated['difference_column'] = df_dated['Open'].diff()
adf_test(df_dated['difference_column'],title='Open Stock Data Differenced');
```

```python
# # # Foreasting with ARIMA (010)(100)[12]
"""
ARIMA(p,d,q)(P,D,Q)[S]

p: The order of the autoregressive (AR) component.
d: The degree of differencing.
q: The order of the moving average (MA) component.
P: The seasonal order of the seasonal autoregressive (SAR) component.
D: The seasonal degree of differencing.
Q: The seasonal order of the seasonal moving average (SMA) component.
S: The seasonal period.

"""
# Fiting model
model = ARIMA(order=(0, 1, 0), seasonal_order=(1, 0, 0, 12))
model.fit(df_train['Open'].to_numpy())

# Making predictions
df_test["ARIMAX_preds"], conf_int = model.predict(n_periods=len(df_test), return_conf_int=True)

# Calculate standard deviation
sigma = (conf_int[:, 1] - conf_int[:, 0]) / (2 * 1.96)

# Evalutaing Model
print('ARIMA Model Summary:')
print(model.summary())
print('RMSE:', np.sqrt(mean_squared_error(np.exp(df_test['Open']), np.exp(df_test['ARIMAX_preds']))))
print('MAE:', mean_absolute_error(np.exp(df_test['Open']), np.exp(df_test['ARIMAX_preds'])))

# Plotting ARIMA predictions
plt.figure(figsize=(12, 5))
_ = plt.plot(df_train['Open'], label = 'Open')
_ = plt.plot(df_test['ARIMAX_preds'], label = 'ARIMAX_preds')
_ = plt.plot(df_test['Open'], label = 'Prev Open')
_ = plt.title("Forecast from ARIMA(0,1,0)(1,0,0)[12]", fontsize=24)
_ = plt.xlabel("Year")
_ = plt.ylabel("Log Price")
_ = plt.legend(loc='upper left');

#  Plotting the confidence interval for the predictions
forecast = df_test["ARIMAX_preds"]
_ = plt.fill_between(df_test.index, forecast - 1.96 * sigma, forecast + 1.96 * sigma, color='orange', alpha=0.4)
_ = plt.fill_between(df_test.index, forecast - 2.75 * sigma, forecast + 2.75 * sigma, color='orange', alpha=0.2)

# Converting the logarithmic data back to normal format
df_comparison = pd.DataFrame()
df_comparison['Actual Stock Value'] = np.exp(df_test['Open'])
df_comparison['Forecasted Nifty'] = np.exp(df_test['ARIMAX_preds'])
df_comparison['residual'] = df_comparison['Actual Stock Value'] - df_comparison['Forecasted Nifty']

# Perform Ljung-Box test
lb_test_results = acorr_ljungbox(df_comparison.residual, lags=[10])
p_value, lag = lb_test_results['lb_pvalue'][10], 10
if p_value < 0.05:
    print(f'Ljung-Box test at lag {lag}: Autocorrelation is significant (p-value = {p_value})')
else:
    print(f'Ljung-Box test at lag {lag}: Autocorrelation is not significant (p-value = {p_value})')

'''L-Jung Box test shows autocorrealtion within the residuals of our data'''

# Measure percentage error
# df_test = np.exp(df_test)
difference = np.abs((np.exp(df_test['Open']) - np.exp(df_test['ARIMAX_preds']))/np.exp(df_test['Open']))
# print(difference)
print("Mean percentage error", np.mean(difference))

np.mean(np.exp(sigma))
```

```python
# # # LSTM for forecasting
# Loading and preprocessing dataset
df_forc = pd.read_csv("/content/NIFTY 500.csv", parse_dates=['Date'], index_col='Date')
df_forc.interpolate(method='time', inplace=True)
df_forc["Open"] = np.log(df_forc["Open"])
df_train = df_forc["2012" : "2015"]
df_test = df_forc["2016"]
train_data = df_train['Open'].values.reshape(-1, 1) # (993, 1)
test_data = df_test['Open'].values.reshape(-1, 1) # (247, 1)
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(train_data)
test_data_scaled = scaler.transform(test_data)

# Prepare data to use vectors of 247 data points for training
X, y = [], []
for i in range(len(data_scaled) - 494):
    X.append(data_scaled[i:i+247, 0])
    y.append(data_scaled[i+247:i+494, 0])
X, y = np.array(X), np.array(y)

# Define model
model = Sequential()
model.add(LSTM(50, input_shape=(247, 1)))
model.add(Dense(247))
model.compile(optimizer='adam', loss='mse')

# Reshape X for LSTM [samples, timesteps, features]
X = X.reshape(X.shape[0], X.shape[1], 1)


plt.figure(figsize=(12, 5))
result_matrix = []
for g in range(50):
  print(f'Stage: {g}\n------------------------------------------------------------------------')
  # LSTM prediction
  model.fit(X, y, epochs=20, batch_size=32, verbose=2)
  last_window = test_data_scaled[:247].reshape(1, 247, 1)
  forecast = model.predict(last_window)
  # Tranform prediciton into original format
  forecast_original_scale = scaler.inverse_transform(forecast)
  forecast_index = df_test.index[-247:]  # предполагаем, что df_test содержит хотя бы 12 последних записей
  forecast_series = pd.Series(forecast_original_scale[0], index=forecast_index)
  # Store prediction
  result_matrix.append(forecast_series)
```

```python
# Plotting the result
plt.figure(figsize=(12, 5))
result_matrix = np.array(result_matrix)
# for i in range(len(result_matrix)):
#   _ = plt.plot(forecast_series.index, np.exp(result_matrix[i]), color='green', alpha=0.1)

# Calculating mean
mean_result = np.mean(result_matrix, axis=0)
# Plotting data
_ = plt.plot(forecast_series.index, np.exp(mean_result), color='red', alpha =0.8)
_ = plt.plot(df_test.index, np.exp(df_test['Open']), label='Actual Open Price', color='blue')
_ = plt.plot(df_train.index, np.exp(df_train['Open']), label='Prev Open', color='black')
_ = plt.title('LSTM Time Series Forecasting')
_ = plt.xlabel('Date')  # Изменил 'Year' на 'Date', так как это может быть более точным описанием оси
_ = plt.ylabel('Open Price')
_ = plt.legend()

# Calculating variance
sigma2 = np.std(np.exp(result_matrix), axis=0)
# PLotting variance
_ = plt.fill_between(df_test.index, np.exp(mean_result)  - 1.96 * sigma2, np.exp(mean_result) + 1.96 * sigma2, color='orange', alpha=0.4)
_ = plt.fill_between(df_test.index, np.exp(mean_result) - 2.75 * sigma2, np.exp(mean_result) + 2.75 * sigma2, color='orange', alpha=0.2)
plt.show()

# Evalutaing Model
print('LSTM Model Summary:')
print(model.summary())
print('RMSE:', np.sqrt(mean_squared_error(np.exp(df_test['Open']), np.exp(mean_result))))
print('MAE:', mean_absolute_error(np.exp(df_test['Open']), np.exp(mean_result)))

# Measure percentage error
difference = np.abs((np.exp(df_test['Open']) - np.exp(mean_result))/np.exp(df_test['Open']))
# print(difference)
print("Mean percentage error", np.mean(difference))
```
